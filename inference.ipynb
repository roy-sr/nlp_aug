# Input Dataset

```python
import pandas as pd

indf = pd.read_csv('dataset/labeled_messages.csv', header=0)
indf.drop('ignore', 1, inplace=True)
indf.index.name = 'id'
indf.replace(r"^ +| +$", r"", regex=True, inplace=True)
indf.head()
```

```python
len(indf)
```

```python
indf_x1 = indf[['empathy']].reset_index().groupby(['empathy']).agg([ 'count'])

pd.set_option('display.max_rows', indf_x1.shape[0]+1)
indf_x1
```

```python
indf_x2 = (indf.astype(str).assign(empathy=indf.empathy.str.split(','))
                          .explode('empathy').reset_index())
indf_x2 = indf_x2[['id','empathy']].groupby(['empathy']).agg([ 'count'])
pd.set_option('display.max_rows', indf_x2.shape[0]+1)
indf_x2
```

# Output dataset

```python
import pandas as pd

outdf = pd.read_csv('dataset/new_labeled_messages.csv', header=0, index_col='index')
outdf.index.name = 'id'
outdf.replace(r"^ +| +$", r"", regex=True, inplace=True)
outdf.head()
```

```python
len(outdf) -len(indf)
```

```python
outdf_x1 = outdf[['empathy']].reset_index().groupby(['empathy']).agg([ 'count'])

pd.set_option('display.max_rows', outdf_x1.shape[0]+1)
outdf_x1
```

```python
outdf_x2 = (outdf.astype(str).assign(empathy=outdf.empathy.str.split(','))
                          .explode('empathy').reset_index())
outdf_x2 = outdf_x2[['id','empathy']].groupby(['empathy']).agg([ 'count'])
pd.set_option('display.max_rows', outdf_x2.shape[0]+1)
outdf_x2
```
